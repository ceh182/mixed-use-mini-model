# PS4 / PS5 Mixed-Use NOI Visualization (Colab Template)

Use this notebook to create a **stacked NOI chart** for your Burnham Place mixed-use program.

```python
import matplotlib.pyplot as plt

# Enter your NOI values from Excel here
residential_noi = 13626563
office_noi = 988000
retail_noi = 1443750
hotel_noi = 585000

labels = ['Burnham Place']
res = [residential_noi]
off = [office_noi]
ret = [retail_noi]
hot = [hotel_noi]
```

```python
# Build stacked NOI chart
plt.figure(figsize=(8,6))
plt.bar(labels, res, label='Residential NOI')
plt.bar(labels, off, bottom=res, label='Office NOI')
plt.bar(labels, ret, bottom=[r+o for r,o in zip(res, off)], label='Retail NOI')
plt.bar(labels, hot, bottom=[r+o+t for r,o,t in zip(res, off, ret)], label='Hotel NOI')
plt.ylabel('Annual NOI ($)')
plt.title('Burnham Place Mixed-Use NOI Stack')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Save chart as PNG
out_path = 'burnham_place_noi_stack.png'
plt.figure(figsize=(8,6))
plt.bar(labels, res, label='Residential NOI')
plt.bar(labels, off, bottom=res, label='Office NOI')
plt.bar(labels, ret, bottom=[r+o for r,o in zip(res, off)], label='Retail NOI')
plt.bar(labels, hot, bottom=[r+o+t for r,o,t in zip(res, off, ret)], label='Hotel NOI')
plt.ylabel('Annual NOI ($)')
plt.title('Burnham Place Mixed-Use NOI Stack')
plt.legend()
plt.tight_layout()
plt.savefig(out_path, dpi=200)
out_path
```
